```python
import os
import sys
import numpy as np
import pandas as pd
import hdf5_getters
```

```python
def write_songs(hdf5paths):
    getters = [x for x in list(hdf5_getters.__dict__.keys()) if x[:4] == 'get_']
    getters.remove("get_num_songs") # special case
    getters = np.sort(getters)
    data = {}
    for getter in getters:
        data[getter[4:]] = []
    for hdf5path in hdf5paths:
        h5 = hdf5_getters.open_h5_file_read(hdf5path)
        for getter in getters:    
            res = hdf5_getters.__getattribute__(getter)(h5)
            data[getter[4:]].append(res)
        h5.close()
    return data
```

```python
def read_dir(directory_str):
    directory = os.fsencode(directory_str)
    songs = np.array([])
    for file in os.listdir(directory):
        path = os.path.join(directory_str, os.fsdecode(file))
        if(os.path.isdir(path)):
            songs = np.append(songs, read_dir(path))
        elif(path.endswith(".h5")):
            songs = np.append(songs, path)
    return songs
```

```python
directory_str = os.path.join('..', 'MillionSongSubset', 'data', 'A', 'D') # You can try MillionSongSubset/data/ only if you want all data (bit slow)
a = read_dir(directory_str)
data = write_songs(a)
df = pd.DataFrame(data)
```
